# Data Cleaning Pipeline (REWE Data)

```python
import pandas as pd
```

```python
df = pd.read_csv("rewe.csv")
print(df.to_string())
```

## Filter and then create category "brand"

```python
product_names = df[df["vacuum"].str.isupper() & df["vacuum"].str.isalpha() & df["size"].isna()]["vacuum"]
df["brand"] = product_names
df["brand"] = df["brand"].ffill()
df = df[~df["size"].isna()]
```

```python
df.to_csv("rewe-cleaned.csv",index=False)
```
